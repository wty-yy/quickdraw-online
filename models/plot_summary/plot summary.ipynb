```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pathlib import Path
plt.rcParams['font.family'] = ['serif']
```

```python
path = Path("./accuracy")
train_dc = pd.read_csv(path.joinpath("deeper_cnn_train.csv"))
val_dc = pd.read_csv(path.joinpath("deeper_cnn_val.csv"))
```

```python
train_cnn = pd.read_csv(path.joinpath("cnn_train_1.csv"))
train_cnn = pd.concat([train_cnn, pd.read_csv(path.joinpath("cnn_train_2.csv"))])
train_cnn = train_cnn.reset_index(drop=True)['Value']
val_cnn = pd.read_csv(path.joinpath("cnn_val_1.csv"))
val_cnn = pd.concat([val_cnn, pd.read_csv(path.joinpath("cnn_val_2.csv"))])
val_cnn = val_cnn.reset_index(drop=True)['Value']
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(train_dc['Step']+1, train_dc['Value'], '*-', label="Deeper CNN train")
ax.plot(val_dc['Step']+1, val_dc['Value'], '*-', label="Deeper CNN val", ls='-')

ax.plot(train_cnn.index+1, train_cnn, '.-', label="CNN train", ls='-')
ax.plot(val_cnn.index+1, val_cnn, '.-', label="CNN val", ls='-')

ax.grid('show', alpha=0.5)
ax.set_ybound(0.05,0.95)
ax.set_xbound(1, 40)
ax.legend(loc='lower right')
ax.set_title("Accuracy(40 Step)")
fig.tight_layout()
plt.savefig("accuracy.png", dpi=100)
plt.show()
```
